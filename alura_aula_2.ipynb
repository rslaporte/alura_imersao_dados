# **Aula 2 - Estatísticas, visualização de dados e distribuições**

```python
import pandas as pd

url_data = 'https://github.com/alura-cursos/imersaodados3/blob/main/dados/dados_experimentos.zip?raw=true'

data = pd.read_csv(url_data, compression = 'zip')
```

```python
map = {"droga" : "composto"}
data.rename(columns=map, inplace=True)
data
```

```python
cod_composto = data['composto'].value_counts().index[:5]
cod_composto
```

--- Making Queries

```python
data.query('composto in @cod_composto')
```

---- Ploting graphs with seaborn and matplotlib

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(8,6))

sns.set()
ax = sns.countplot(x = 'composto', data=data.query('composto in @cod_composto'))
ax.set_title('Top 5 Compostos')
plt.show()
```
